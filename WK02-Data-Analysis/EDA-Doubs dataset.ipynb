## 探索式資料分析 統計圖表
### The Doubs dataset have been collected at 30 sites along the Doubs River (near the France–Switzerland border in the Jura Mountains. )
#### DoubsAbu: contains coded abundances (豐富充足) of 27 fish species.
#### DoubsEnv: contains 11 environmental variables related to the hydrology, geomorphology and chemistry of the river.
#### DoubsSpa: contains the geographical coordinates (Cartesian, X and Y) of the sites.

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df_Abu = pd.read_csv('DoubsAbu.csv')
print(df_Abu.shape)
df_Abu.head()
```

```python
df_Env = pd.read_csv('DoubsEnv.csv')
print(df_Env.shape)
df_Env.head()
```

```python
df_Spa = pd.read_csv('DoubsSpa.csv')
print(df_Spa.shape)
df_Spa.head()
```

```python
abundance_class = []
for col in df_Abu:
    abundance_class.append(df_Abu[col].value_counts())
abundance_class = pd.DataFrame(abundance_class)

abundance_class.sum(axis=0).plot(kind='bar')
plt.xlabel('Abundance class')
plt.ylabel('Frequency')
plt.xticks(rotation=0)
plt.tight_layout()
```

```python
plt.plot(df_Spa['x'], df_Spa['y'])

for i in range(df_Spa.shape[0]):
    plt.annotate(str(i+1), (df_Spa['x'][i], df_Spa['y'][i]))

plt.xlabel('x coordinate (km)')
plt.ylabel('y coordinate (km)')
plt.title('Site Locations')
```

```python
plt.plot(df_Spa['x'], df_Spa['y'])

for i in range(df_Spa.shape[0]):
    plt.scatter(df_Spa['x'][i], df_Spa['y'][i], marker='o', s=np.square(df_Abu.iloc[i]['TRU'])*20, facecolor='w', edgecolor='k')

plt.xlabel('x coordinate (km)')
plt.ylabel('y coordinate (km)')
plt.title('Brown Trout')
```

```python
# Compute the number of sites where each species is present
species_occurance = (df_Abu > 0).astype(int).sum(axis=0).sort_values()

species_occurance.plot(kind='hist', bins=np.linspace(0, 30, 7), edgecolor='w')
plt.xlabel('Number of Occurance')
plt.ylabel('Number of Species')
plt.show()
species_occurance = species_occurance / 30 * 100
species_occurance.plot(kind='hist', bins=np.linspace(0, 100, 11), edgecolor='w')
plt.xlabel('Frequency of Occurance (%)')
plt.ylabel('Number of Species')
```

```python
species_richness = (df_Abu > 0).astype(int).sum(axis=1)

plt.plot(np.arange(1, df_Abu.shape[0]+1), species_richness)
for i in range(df_Spa.shape[0]):
    plt.annotate(str(i+1), (i+1, species_richness[i]))

plt.xlabel('Position of sites along the river')
plt.ylabel('Species richness')
plt.title('Species richness vs.\nUpstream-Downstream Gradient')
plt.show()

plt.plot(df_Spa['x'], df_Spa['y'], c='b')

for i in range(df_Spa.shape[0]):
    plt.scatter(df_Spa['x'][i], df_Spa['y'][i], marker='o', s=np.square(species_richness[i]), c='indianred', edgecolor='b')

plt.xlabel('x coordinate (km)')
plt.ylabel('y coordinate (km)')
plt.title('Map of species richness')
```

```python
plt.boxplot([df_Abu['LOC'], np.sqrt(df_Abu['LOC']), np.log(df_Abu['LOC'] + 1)])
plt.xticks(np.arange(1, 4), ['raw data', 'sqrt', 'log1p'])
plt.title('Simple Transformation')
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 4))
axes = axes.ravel()
cols = ['alt', 'deb', 'oxy', 'nit']
colors = ['red', 'green', 'blue', 'brown']
titles = ['Altitude', 'Discharge', 'Oxygen', 'Nitrate']

marker_size = df_Env.copy()
for col in df_Env:
    Vmin = df_Env[col].min()
    Vmax = df_Env[col].max()
    marker_size[col] = (df_Env[col] - Vmin) / (Vmax - Vmin)
    
for (col, title, color, ax) in zip(cols, titles, colors, axes):
    ax.plot(df_Spa['x'], df_Spa['y'], c='b')
    for i in range(df_Spa.shape[0]):
        ax.scatter(df_Spa['x'][i], df_Spa['y'][i], marker='o', s=marker_size[col][i] * 400, c=color, edgecolor='k')
        ax.set_xlabel('x coordinate (km)')
        ax.set_ylabel('y coordinate (km)')
        ax.set_title(title)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 4))
axes = axes.ravel()
cols = ['alt', 'deb', 'oxy', 'nit']
colors = ['red', 'green', 'blue', 'brown']
ylabels = ['Altitude (m)', 'Discharge (m3/s)', 'Oxygen (mg/L)', 'Nitrate (mg/L)']
titles = ['Altitude', 'Discharge', 'Oxygen', 'Nitrate']

for (col, title, ylabel, color, ax) in zip(cols, titles, ylabels, colors, axes):
    ax.plot(df_Env['das'], df_Env[col], c=color)
    ax.set_xlabel('Distance from the source (km)')
    ax.set_ylabel(ylabel)
    ax.set_title(title)

plt.tight_layout()
```

```python
#河道坡度
fig, axes = plt.subplots(1, 4, figsize=(15, 4))
axes = axes.ravel()

axes[0].hist(df_Env['pen'], bins=np.linspace(0, 50, 6), edgecolor='w')
axes[0].set_xlabel('Env[pen]')
axes[0].set_ylabel('Frequency')

axes[1].hist(np.log(df_Env['pen']), bins=np.linspace(-2, 4, 7), edgecolor='w')
axes[1].set_xlabel('ln(Env[pen])')
axes[1].set_ylabel('Frequency')

axes[2].boxplot(df_Env['pen'])
axes[2].set_xticklabels(['Env[pen]'])

axes[3].boxplot(np.log(df_Env['pen']))
axes[3].set_xticklabels(['ln(Env[pen])'])

plt.tight_layout()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

fig, axes = plt.subplots(2, 2, figsize=(8, 6))

for i, ax in enumerate(axes.ravel()):
    ax.hist(iris.data[:, i], bins=np.linspace(0, 8, 17), edgecolor='w')
    ax.set_ylabel('Frequency')
    ax.set_title(iris.feature_names[i])
plt.tight_layout()
```
